```python
# print initial words
print ('Welcome to ROCK, PAPER, SCISSORS game!')

# counter for "win", "lose", and "tie"
counter_win = 0
counter_lose = 0
counter_tie = 0

# while loop for playing again and again till win,
# which means the "counter_win add to 1."
# the condition is win or one should play again
while counter_win != 1:
    
    # user input
    # "compare" which is used to compare with AI(computer) get the user's input  
    # "com" get a corresponding string as a user inputs 'r', 'p', or 's' 
    # while there is a wrong input, print the hint "It's not a correct form!"
    compare = input ('Enter your move: (r)ock (p)aper (s)cissors')
    if compare == 'r':
        com = "ROCK"
    elif compare == 'p':
        com = "PAPER"
    elif compare == 's':
        com = "SCISSORS"
    else:
        print("It's not a correct form!")
        continue
    
    # "ai" for getting a random number
    # 'a' get a corresponding string when random generating a number
    # random generate ai and give it a corresponding string
    import random
    ai = random.randint(1, 3)
    if ai == 1:
        a = "ROCK"
    elif ai == 2:
        a = "PAPER"
    else:
        a = "SCISSORS"
    
    # print the versus sentences
    print (f'{com}  versus...\n{a}')
    
    # compare strings
    # judge the conditions "win", "lose", and "tie"
    # counter++ at each condition
    if com == a:
        print("It is a tie!")
        counter_tie = counter_tie + 1
    elif com == 'ROCK' and a == 'PAPER':
        print("You lose!")
        counter_lose = counter_lose + 1
    elif com == 'ROCK' and a == 'SCISSORS':
        print("You win!")
        counter_win = counter_win + 1
    elif com == 'PAPER' and a == 'ROCK':
        print("You win!")
        counter_win = counter_win + 1
    elif com == 'PAPER' and a == 'SCISSORS':
        print("You lose!")
        counter_lose = counter_lose + 1
    elif com == 'SCISSORS' and a == 'PAPER':
        print("You win!")
        counter_win = counter_win + 1
    elif com == 'SCISSORS' and a == 'ROCK':
        print("You lose!")
        counter_lose = counter_lose + 1

# print out the final answer
print (f"You have {counter_tie} ties and {counter_lose} losses before your first win.")
```
